## üß† Training Process in Machine Learning
Training is the process where a machine learning model learns patterns from data so it can make predictions.

‚öôÔ∏è Step-by-Step Breakdown:
- 1Ô∏è‚É£ Dataset Preparation
You start with a training dataset (input features + labels).

 - The data is usually preprocessed ‚Äî cleaned, normalized, and split into training and testing sets.

- 2Ô∏è‚É£ Batching
Instead of feeding the entire dataset to the model at once, we split it into small groups called batches.

 - Example: If your batch size is 32, the model sees 32 records at a time.

## üìå Why use batches?

- Saves memory

- Makes training faster and smoother

- Allows for gradient updates after each batch

## 3Ô∏è‚É£ Epochs
An epoch is one full pass through the entire training dataset.

If you set epochs = 10, the model will see all data 10 times.

### üìå Purpose of multiple epochs?
So the model can learn better by seeing the data repeatedly and improving its predictions.

###4Ô∏è‚É£ Forward Pass
For each batch, the model makes predictions.

It calculates how far off the predictions are using a loss function.

###5Ô∏è‚É£ Backward Pass (Backpropagation)
The model adjusts its internal weights using gradients (from the loss).

This is done by an algorithm called Gradient Descent.

###6Ô∏è‚É£ Repeat
This process of:

forward pass ‚Üí loss calculation ‚Üí backpropagation ‚Üí weight update
is repeated for every batch, and for every epoch.

###7Ô∏è‚É£ Evaluation
After training, the model is tested on unseen testing data to evaluate how well it generalizes.

üß™ Example:
If you have:

1000 training samples

Batch size = 100

Epochs = 5

‚û°Ô∏è Then the model will do:

10 batches per epoch

5 epochs √ó 10 batches = 50 training steps

## ‚úÖ Step 7: Input Function + Batching Data (Updated for TensorFlow 2.x+)

üß† Concept Summary
- Input function: A wrapper that returns a tf.data.Dataset object.

- Batching: The dataset is divided into smaller chunks (batch_size), e.g., 32.

- Epochs: One full pass over the dataset. repeat(n) does this n times.

- Shuffling: Useful in training to randomize data and avoid learning patterns in order.



Your DataFrame is like a full pizza üçï

The model can‚Äôt eat the whole thing at once...

So you slice it into small pieces (batches üçïüçï), and feed one slice at a time.

Also, you mix up the slices (shuffle) so the model doesn‚Äôt learn in order.

And you repeat this for multiple meals (epochs).

### ‚ùì What is an input function?

It‚Äôs a function that prepares and feeds data into the model during training or evaluation.

‚úÖ Key Points:

- It converts data (features and labels) into a tf.data.Dataset object.

- It can shuffle the data (for randomness) and repeat it (for multiple epochs).

- It returns batches of data (usually a batch size like 32, 64, etc.).


### ‚ùì What are feature columns?
Feature columns tell TensorFlow how to interpret and use input features.

‚úÖ Why do we need them?

Converts raw input (like text, numbers, categories) into a format the model can understand.

Helps TensorFlow know if a column is numeric, categorical, or needs special processing.

```python
#  Example Code:

def input_fn(features, labels, training=True, batch_size=256):
    # Convert inputs to tf.data.Dataset
    dataset = tf.data.Dataset.from_tensor_slices((features, labels))

    # Shuffle and repeat during training
    if training:
        dataset = dataset.shuffle(1000).repeat()

    # Return batches of data
    return dataset.batch(batch_size)


# Let's say we have numeric features: 'sepal_length', 'sepal_width'
for header in ['sepal_length', 'sepal_width']:
    feature_columns.append(tf.feature_column.numeric_column(header))
```

```python
def make_input_fn(data_df, label_df, num_epochs=10, shuffle=True, batch_size=32):
    def input_function():
        # 1. Turn the DataFrame into a TensorFlow Dataset
        ds = tf.data.Dataset.from_tensor_slices((dict(data_df), label_df))

        # 2. Shuffle it if we‚Äôre training
        if shuffle:
            ds = ds.shuffle(buffer_size=1000)

        # 3. Split into batches, and repeat for multiple epochs
        ds = ds.batch(batch_size).repeat(num_epochs)

        return ds
    return input_function
```

## ‚úÖ Step 8: Creating the Model (Linear Estimator)

Since we're using the current TensorFlow version (2.14+ or higher), Estimators are deprecated by default.

So instead of tf.estimator.LinearClassifier, we‚Äôll now use Keras, which is TensorFlow‚Äôs modern and recommended high-level API.

```python
import tensorflow as tf
import pandas as pd

# Load Titanic dataset
dftrain = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/train.csv')
dfeval = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/eval.csv')
y_train = dftrain.pop('survived')
y_eval = dfeval.pop('survived')
```

### ‚öôÔ∏è Step 1: Preprocess data (normalize & encode)

```python
# One-hot encode categorical features
dftrain = pd.get_dummies(dftrain)
dfeval = pd.get_dummies(dfeval)

# Match columns in case test set misses some categories
dftrain, dfeval = dftrain.align(dfeval, join='left', axis=1, fill_value=0)
```

### üß™ Step 2: Create TensorFlow datasets

```python
def df_to_dataset(data, labels, shuffle=True, batch_size=32):
    ds = tf.data.Dataset.from_tensor_slices((dict(data), labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(data))
    ds = ds.batch(batch_size)
    return ds

batch_size = 32
train_ds = df_to_dataset(dftrain, y_train, batch_size=batch_size)
eval_ds = df_to_dataset(dfeval, y_eval, shuffle=False, batch_size=batch_size)
```

### üß± Step 3: Build Keras model

```python
!pip install --upgrade tensorflow

```

```python
from tensorflow.keras import layers

# Input layer: create Input() for each feature
inputs = []
for name in dftrain.columns:
    inputs.append(tf.keras.Input(shape=(1,), name=name))

# Concatenate all inputs
concatenated = layers.concatenate(inputs)

# Hidden layer + output
x = layers.Dense(16, activation='relu')(concatenated)
output = layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(inputs=inputs, outputs=output)
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(train_ds, epochs=10)

loss, accuracy = model.evaluate(eval_ds)
print(f"Evaluation accuracy: {accuracy:.2f}")
```

## ‚úÖ Step 9: Making Predictions with the Trained Model (Titanic)

```python
# Get a few samples from evaluation data
sample = dfeval.iloc[:10]  # First 10 passengers

# Convert each column to tensor and store them in a list
sample_tensors = [tf.convert_to_tensor(sample[column]) for column in sample.columns]

# Get predictions
predictions = model.predict(sample_tensors)

# Print survival probabilities
for i, p in enumerate(predictions.flatten()):
    print(f"Passenger {i+1} - Survival Probability: {p:.2f}")

import matplotlib.pyplot as plt
plt.hist(predictions, bins=10)
plt.title("Predicted Survival Probabilities")
plt.xlabel("Probability of Survival")
plt.ylabel("Number of Passengers")
plt.show()
```

## ‚úÖ Step 10: Classification (Iris Dataset ‚Äì 3 classes)

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

iris = load_iris()
X = iris['data']
y = iris['target']  # 3 classes: 0, 1, 2

# Split and scale
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### üß† Build a Keras Classifier

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model_iris = Sequential([
    Dense(10, activation='relu', input_shape=(4,)),
    Dense(10, activation='relu'),
    Dense(3, activation='softmax')  # 3 classes
])
```

### ‚öôÔ∏è Compile + Train

```python
model_iris.compile(optimizer='adam',
                   loss='sparse_categorical_crossentropy',
                   metrics=['accuracy'])

model_iris.fit(X_train, y_train, epochs=20)
```

### ‚úÖ Evaluate + Predict

```python
loss, acc = model_iris.evaluate(X_test, y_test)
print(f"Iris Classification Accuracy: {acc:.2f}")

# Predict class for one flower
sample = np.array([X_test[0]])
pred = model_iris.predict(sample)
print("Predicted class:", np.argmax(pred))
```

## ‚úÖ Step 12: Hidden Markov Models (HMM) ‚Äî with tensorflow_probability

üéØ What is a Hidden Markov Model?
An HMM is used when:

- You have hidden states (e.g., weather is cold/hot)

- And observable outcomes (e.g., mood or temperature)

- You want to model the probability of future events over time

We'll model a simple weather system.

```python
import tensorflow as tf
import tensorflow_probability as tfp
tfd = tfp.distributions

# Step 3: Define the HMM Components

# Initial probabilities: 80% cold, 20% hot
initial_distribution = tfd.Categorical(probs=[0.8, 0.2])

# Transition matrix: [cold‚Üícold, cold‚Üíhot], [hot‚Üícold, hot‚Üíhot]
transition_distribution = tfd.Categorical(probs=[
    [0.7, 0.3],  # If today is cold
    [0.2, 0.8]   # If today is hot
])

# Observation distribution: temperatures per state
observation_distribution = tfd.Normal(loc=[0., 15.], scale=[5., 10.])

# Step 4: Create the HMM Model

model = tfd.HiddenMarkovModel(
    initial_distribution=initial_distribution,
    transition_distribution=transition_distribution,
    observation_distribution=observation_distribution,
    num_steps=7  # Predict 7 days
)

# Step 5: Compute Mean Temperatures (Expected Output)

mean = model.mean()

# In TensorFlow 2.x, eager execution is enabled by default, so you can directly compute and print the result
print("Expected temperatures over 7 days:\n", mean.numpy())  # Use .numpy() to get the value
```

##üß† Code: Simple Neural Network with One Hidden Layer

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# 1. Load MNIST data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Normalize pixel values (0‚Äì255 ‚Üí 0‚Äì1)
x_train, x_test = x_train / 255.0, x_test / 255.0

# 3. One-hot encode the labels (0‚Äì9 ‚Üí [0,0,0,1,0,...])
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# 4. Create the model
model = Sequential([
    Flatten(input_shape=(28, 28)),       # Flatten 28x28 images to 784
    Dense(128, activation='relu'),       # 1st hidden layer
    Dense(10, activation='softmax')      # Output layer (10 classes)
])

# 5. Compile the model
model.compile(
    optimizer='adam',                     # Optimizer
    loss='categorical_crossentropy',     # Loss function
    metrics=['accuracy']                 # Metric to track
)

# 6. Train the model
model.fit(x_train, y_train, epochs=5, batch_size=32)

# 7. Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
IMG_SIZE = 160

def format_example(image, label):
    image = tf.cast(image, tf.float32)
    image = (image / 127.5) - 1  # normalize to [-1, 1]
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    return image, label

train = raw_train.map(format_example).shuffle(1000).batch(32)
validation = raw_validation.map(format_example).batch(32)
test = raw_test.map(format_example).batch(32)
```

25,000 preprocessed movie reviews.

```python
# ========== SHAKESPEARE TEXT GENERATION FULL FLOW ==========
import tensorflow as tf
import numpy as np
import os

# ========== STEP 1: Download and Load Shakespeare Data ==========
url = "https://storage.googleapis.com/download.tensorflow.org/data/shakespeare.txt"
path_to_file = tf.keras.utils.get_file('shakespeare.txt', url)

# Read the file
text = open(path_to_file, 'rb').read().decode(encoding='utf-8')
print(f"Length of text: {len(text)} characters")

# ========== STEP 2: Create Character Vocabulary ==========
vocab = sorted(set(text))
print(f"{len(vocab)} unique characters")

# Create mappings
char2idx = {char: idx for idx, char in enumerate(vocab)}
idx2char = np.array(vocab)
text_as_int = np.array([char2idx[c] for c in text])

# ========== STEP 3: Create Sequences ==========
seq_length = 100
examples_per_epoch = len(text) // (seq_length + 1)

char_dataset = tf.data.Dataset.from_tensor_slices(text_as_int)
sequences = char_dataset.batch(seq_length + 1, drop_remainder=True)

def split_input_target(chunk):
    input_text = chunk[:-1]
    target_text = chunk[1:]
    return input_text, target_text

dataset = sequences.map(split_input_target)

# ========== STEP 4: Build Dataset Batches ==========
BATCH_SIZE = 64
BUFFER_SIZE = 10000

dataset = (
    dataset.shuffle(BUFFER_SIZE)
    .batch(BATCH_SIZE, drop_remainder=True)
)

# ========== STEP 5: Model Configuration ==========
vocab_size = len(vocab)
embedding_dim = 256
rnn_units = 1024

def build_model(vocab_size, embedding_dim, rnn_units, batch_size):
    return tf.keras.Sequential([
        tf.keras.layers.Input(batch_shape=(batch_size, None)),
        tf.keras.layers.Embedding(vocab_size, embedding_dim),
        tf.keras.layers.LSTM(rnn_units, return_sequences=True, stateful=True),
        tf.keras.layers.Dense(vocab_size)
    ])

# Build training model
model = build_model(vocab_size, embedding_dim, rnn_units, BATCH_SIZE)

# ========== STEP 6: Compile Model ==========
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
)

# ========== STEP 7: Create Checkpoint Callback ==========
checkpoint_dir = './training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, 'gen_model.weights.h5')
os.makedirs(checkpoint_dir, exist_ok=True)

checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_prefix,
    save_weights_only=True,
    save_best_only=True,
    verbose=1
)

# ========== STEP 8: Train Model ==========
EPOCHS = 5  # Can increase to 20+ for better output

model.fit(
    dataset,
    epochs=EPOCHS,
    callbacks=[checkpoint_callback]
)

# ========== STEP 9: Load Trained Weights into Inference Model ==========
# Use batch_size = 1 for inference
gen_model = build_model(vocab_size, embedding_dim, rnn_units, batch_size=1)

# Load weights
if os.path.exists(checkpoint_prefix):
    gen_model.load_weights(checkpoint_prefix)
    gen_model.build(tf.TensorShape([1, None]))
    print("‚úÖ Weights loaded successfully.")
else:
    raise FileNotFoundError(f"‚ùå No weights file found at {checkpoint_prefix}")

# ========== STEP 10: Text Generation Function ==========
def generate_text(model, start_string):
    num_generate = 500  # Number of characters to generate
    temperature = 1.0   # Higher = more random; lower = more predictable

    # Vectorize input
    input_eval = tf.expand_dims([char2idx[s] for s in start_string], 0)
    text_generated = []

    # Reset model states
    for layer in model.layers:
        if hasattr(layer, 'reset_states'):
            layer.reset_states()

    # Generate characters
    for _ in range(num_generate):
        predictions = model(input_eval)
        predictions = tf.squeeze(predictions, 0)
        predictions = predictions / temperature

        predicted_id = tf.random.categorical(predictions, num_samples=1)[-1, 0].numpy()
        input_eval = tf.expand_dims([predicted_id], 0)
        text_generated.append(idx2char[predicted_id])

    return start_string + ''.join(text_generated)

# ========== STEP 11: Generate and Print Sample Text ==========
print("\n========= TEXT GENERATION SAMPLE =========\n")
print(generate_text(gen_model, start_string="ROMEO: "))
```

## üß™ 6. Practical Example ‚Äî FrozenLake-v0
A simple OpenAI Gym environment where:

Agent must reach a goal (G) without falling in a hole (H).

Map: 4x4 grid = 16 states.

4 actions: left, right, up, down.

```python
# ========== Q-LEARNING (FrozenLake) - Updated Version ==========
import gym
import numpy as np
import matplotlib.pyplot as plt

# 1. Environment Setup
env = gym.make('FrozenLake-v1', is_slippery=False)  # Non-slippery for deterministic behavior

STATES = env.observation_space.n
ACTIONS = env.action_space.n
Q = np.zeros((STATES, ACTIONS))

# 2. Training Parameters
EPISODES = 1500
MAX_STEPS = 100
LEARNING_RATE = 0.81
GAMMA = 0.96
epsilon = 0.9
epsilon_min = 0.01
epsilon_decay = 0.001
RENDER = False

rewards = []

# 3. Training Loop
for episode in range(EPISODES):
    reset_result = env.reset()
    state = reset_result[0] if isinstance(reset_result, tuple) else reset_result

    for step in range(MAX_STEPS):
        if RENDER:
            env.render()

        # Œµ-greedy action selection
        if np.random.uniform(0, 1) < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(Q[state, :])

        step_result = env.step(action)
        if len(step_result) == 4:
            next_state, reward, done, _ = step_result
        else:
            next_state, reward, terminated, truncated, _ = step_result
            done = terminated or truncated

        # Q-value update
        Q[state, action] += LEARNING_RATE * (
            reward + GAMMA * np.max(Q[next_state, :]) - Q[state, action]
        )

        state = next_state

        if done:
            rewards.append(reward)
            break

    # Decay epsilon
    epsilon = max(epsilon_min, epsilon - epsilon_decay)

# 4. Print Final Q-table and Performance
print("\nTrained Q-Table:")
print(Q)
print(f"Average reward over all episodes: {np.mean(rewards):.2f}")

# 5. Plot Training Progress
def get_average(values):
    return sum(values) / len(values) if values else 0

avg_rewards = [get_average(rewards[i:i+100]) for i in range(0, len(rewards), 100)]

plt.plot(avg_rewards)
plt.ylabel('Average Reward')
plt.xlabel('Episode Batches (per 100 episodes)')
plt.title('Training Performance over Time')
plt.grid(True)
plt.show()
```
